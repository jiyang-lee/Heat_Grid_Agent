# 08 Model Handoff

이 노트북은 다음 단계 Agent/서비스 팀에 넘길 모델 전달 패키지를 확인하는 용도다.

전달 패키지 위치:

- `model_handoff/heatgrid_ml_models_2026-06-25`


```python
from pathlib import Path
import json
import joblib

ROOT = Path.cwd()
if not (ROOT / 'model_handoff').exists():
    ROOT = Path.cwd().parents[1]

HANDOFF = ROOT / 'model_handoff' / 'heatgrid_ml_models_2026-06-25'
print(HANDOFF)
```

```python
required_files = [
    'anomaly/standard_scaler.joblib',
    'anomaly/isolation_forest.joblib',
    'anomaly/baseline_model_metadata.json',
    'risk/lightgbm_risk_model.joblib',
    'risk/risk_model_group_calibration.json',
    'risk/risk_model_metadata.json',
    'leadtime/lightgbm_leadtime_bucket_model_promoted.joblib',
    'leadtime/leadtime_bucket_model_promoted_metadata.json',
    'priority/priority_engine_tuned_metadata.json',
    'README.md',
    'MANIFEST.json',
]

missing = [path for path in required_files if not (HANDOFF / path).exists()]
if missing:
    raise FileNotFoundError(missing)

print('handoff files ok')
```

```python
joblib.load(HANDOFF / 'anomaly' / 'standard_scaler.joblib')
joblib.load(HANDOFF / 'anomaly' / 'isolation_forest.joblib')
joblib.load(HANDOFF / 'risk' / 'lightgbm_risk_model.joblib')
joblib.load(HANDOFF / 'leadtime' / 'lightgbm_leadtime_bucket_model_promoted.joblib')

for path in HANDOFF.rglob('*.json'):
    json.loads(path.read_text(encoding='utf-8-sig'))

print('model load and json parse ok')
```
